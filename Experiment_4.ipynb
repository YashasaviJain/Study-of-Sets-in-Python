```python
# Set operations
A = {1,2,3,4}
B = {3,4,5,6}
print("Union:",A|B)
print("Intersection:",A&B)
print("Difference1:",A-B)
print("Difference2:",B-A)
print("Symmetric Difference:",A^B)
```

```python
# Find the common subjects chosen by all students using set operations.
student1={"AI","ML","Python","Data Science"}
student2={"Python","ML","Cloud Computing"}
student3={"Python","ML","Dawl"}
common_subjects = student1 & student2 & student3
print("Common Subjects:",common_subjects)
```

```python
# One set contains students in the Cricket Club,another in the Football Club.
# Find students who are:1, In both clubs 2, In only one club
# Concept Used: Intersection, Symmetric Difference
Cricket = {"Amit","Rahul","Sneha"}
Football = {"Rahul","Neha","Sneha"}
print("Both Clubs:",Cricket & Football)
print("Only one Club:", Cricket ^ Football)
```

```python
# Absent Students
# Give a Set of all Students and a set of present students, find the absent Students.
# Concept Used: Set Difference (-)
all_students = {"A","B","C","D","E"}
present_students = {"A","C","E"}
absent_students = all_students - present_students
print("Absent Students:",absent_students)
```

```python
# A set contains valid couce codes.
# Remove a discontinued couse code from the set.
# Concept used: remove() / discard()
courses = {"ETC101","ETC102","ETC103"}
courses.discard("ETC102")
print("Upated Courses:",courses)
```
